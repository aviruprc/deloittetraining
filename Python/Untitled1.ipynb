```python
def group(t1,t2,t3,t4):
    
    #askUser()
    print('enter 1 for team1 and 2 or any other number for team2')
    g1=int(input('who won between {0} and {1}:  '.format(t1,t2)))
    g2=int(input('who won between {0} and {1}:  '.format(t1,t3)))
    g3=int(input('who won between {0} and {1}:  '.format(t1,t4)))
    g4=int(input('who won between {0} and {1}:  '.format(t2,t3)))
    g5=int(input('who won between {0} and {1}:  '.format(t2,t4)))
    g6=int(input('who won between {0} and {1}:  '.format(t3,t4)))
    
    #calculate the score for each team
    t1_score,t2_score,t3_score,t4_score=0,0,0,0
    
    if(g1==1):
        t1_score+=1
    else:
        t2_score+=1
        
    if(g2==1):
        t1_score+=1
    else:
        t3_score+=1
        
    if(g3==1):
        t1_score+=1
    else:
        t4_score+=1
    
    if(g4==1):
        t2_score+=1
    else:
        t3_score+=1
        
    if(g5==1):
        t2_score+=1
    else:
        t4_score+=1
        
    if(g6==1):
        t3_score+=1
    else:
        t4_score+=1
      
    #now this list has the score correspondinf to each team
    lista=[t1_score,t2_score,t3_score,t4_score]
    #print(lista)
    #l=enumerate(lista,1)
    maxpos1 = lista.index(max(lista))  #winning team
    #print(maxpos1)
    if(maxpos1==0):
        win=t1
    elif (maxpos1==1):
        win=t2
    elif (maxpos1==2):
        win=t3
    elif (maxpos1==3):
        win=t4
        
    lista[maxpos1]=0
    
    maxpos2 = lista.index(max(lista))  #2nd pos
    #print(maxpos2)
    if(maxpos2==0):
        second=t1
    elif (maxpos2==1):
        second=t2
    elif (maxpos2==2):
        second=t3
    elif (maxpos2==3):
        second=t4
    
    #print(win,second)
    return win,second
    
```

```python
#pool 1
print("POOL 1 MATCHES")
f1,s1=group('ind','pak','ban','china')

#pool 2
print("POOL 2 MATCHES")
f2,s2=group('japan','vietnam','hongkong','indonesia')

#pool 3
print("POOL 3 MATCHES")
f3,s3=group('myanmar','israel','iran','bhutan')

#pool 4
print("POOL 4 MATCHES")
f4,s4=group('nepal','germany','france','switzerland')

#pool 1 vs pool 2--->pool5
print("MATCHES between POOL 1 AND POOL 2 top 2 teams")
f5,s5=group(f1,s1,f2,s2)

#pool 3 vs pool 4------> pool 6
print("MATCHES between POOL 3 AND POOL 4 top 2 teams")
f6,s6=group(f3,s3,f4,s4)

#pool 5 vs pool 6----->final
print("FINALS:  ")
f7,s7=group(f5,s5,f6,s6)

print("winner:  ", f7)
print("runners up: ",s7)
```
